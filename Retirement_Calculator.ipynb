A perputuity  pays for an infinite amount of time. In finance, perpetuity is a constant stream of identical cash flows with no end. Perhaps you might live for ever in a cryogenic tube and wake up in 2321. this function provides the present value needed to pay for a growing perpuity. 

```python
## cost for retirement depnends on your expenses, inflation, and discount rate.
## calculates the present value of a series of payments that grows each year, forever. 
def retirement(expenses,growth=.03,discount_rate=.08):
  r= expenses/(discount_rate-growth)
  print("Retirement will cost $",r)
```

```python
retirement(60000)
```

```python
## We want to visualize what each payment will be over time. 
## Create 3 lists, period, payments, values. 
```

```python
periods = [] 
t= 0
while t <=500:
  periods.append(t)
  t= t + 1 
```

```python
print(periods)
len(periods)
```

```python
payments = [] ##Create a list for payments 
values = [] ## creates a list for the present value of each payment
```

```python
#Calculates the future value of each payment based on inflation 
for period in periods:
  if period ==0:
    payment= 0 
  else:
    payment= 60000*(1.03)**(period-1)
  payments.append(payment)
```

```python
print(payments)
len(payments)
```

```python
## Calculates the present value of each payment and adds it to the lists
n=0
for payment in payments:
  n=n+1
  v= payment/((1+.08)**(n-1))
  values.append(v)
```

```python
print(values)
len(values)
```

```python
## Adds up all the values in the list. 
sum(values)
```

```python
def retire_plan(start=0,end=500,age= 40,expenses=60_000,growth=.03,discount_rate=.08):
  payments=[]
  values=[]
  n=0
  retire_age= age + start
  dead= age + end
  ##calculates payments list
  for p in range(0,200): ##creates a range of values to calculate payments
    if p == 0:
      payment = 0 
    else:
      payment= expenses*(1+ growth)**(p-1) ## adjusts each payment for inflation
    payments.append(payment)
  ##calculates the values list
  for payment in payments:
    n=n+1
    v= payment/((1+discount_rate)**(n-1))
    values.append(v)
  print(payments)
  print(values)
  cost = (sum(values[start:end + 1]))
  print('you on on track to retire at the age of',retire_age,"if you currently have $",cost,"in savings.")
  print('This plan runs out of money when you are', dead)

  print(len(payments),len(values))
  
```

```python
retire_plan(17,50,expenses=60000)
```
